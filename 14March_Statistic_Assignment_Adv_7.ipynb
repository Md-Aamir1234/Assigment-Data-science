Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio
test. The function should return the F-value and the corresponding p-value for the test.

```python
#Ans1.

# To Calculate the F-value and the correspondin p-value for a variance for a variance ratio test in Python,
# you can use the  'scipy.stats' module which provide functions to perform statistical tests.


import numpy as np
from scipy.stats import f

def variance_ratio_test (data1, data2) :
  # Calculate the variances of the two datasets

  var1 = np.var(data1, ddof=1)
  var2 = np.var(data2, ddof=1)

# Calculate the F-value
  F = var1/var2 if var1 >= var2 else var2/var1

# Calculate the degrees of freedom for two datasets
  dfn = len(data1) -1 # Degrees of freedom for the numerator
  dfn = len(data2) -1 # Degrees of freedom for the denominator

# Calculate the p-value
  p_value = 1 - f.cdf (F, dfn, dfn)

  return F, p_value

# Example usage
data1 = [10,12,14,16,18,20]
data2 = [30,32,34,36,38,40]
F, p_value = variance_ratio_test (data1, data2)
print ("F-value :", F)
print("p-value:", p_value)
```

Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an
F-distribution, write a Python function that returns the critical F-value for a two-tailed test.

```python
#Ans2.

# We can use the inverse cumulative distribution function ('ppf') from the 'scipy.stats' module. for a two-tailed test,
# you need to split the significance level between the two tails  Used Python function to calculate the critical F-value for a two-tailed test:


from scipy.stats import f
def critical_f_value(significance_level, dfn, dfd) :
  # Split the significance level between the two tails
  alpha = significance_level / 2

  # Calculate the critical F-value for the upper tail
  critical_value_upper = f.ppf(1 - alpha, dfn, dfd)

  # Calculate the critical F-value for the lower tail (not used in practice)
  critical_value_lower = f.ppf(1 - alpha, dfn, dfd)

  # Calculate the critical F-value for the lower tail (not used in practice)
  critical_value_lower = f.ppf(alpha, dfn, dfd)

  return critical_value_lower, critical_value_upper

# Example usage
significance_level = 0.05
dfn = 5 # degrees of freedom for the numerator
dfd = 10 # degrees of freedom for the denominator
critical_value_lower, critical_value_upper = critical_f_value(significance_level,dfn, dfd)
("Lower critical F-value:", critical_value_lower)
("Upper critical F-value:", critical_value_upper)

# Conclusion :
# This Function returns the critical F-values for both the lower and upper tails, although the upper critical F-value is typically used to compare against the computed F-value in Hypothesis testing.
```

Q3. Write a Python program that generates random samples from two normal distributions with known
variances and uses an F-test to determine if the variances are equal. The program should output the Fvalue, degrees of freedom, and p-value for the test.

```python
#Ans3.

# Here's a Python program that generates random samples from two normal distributions with known variances and uses an F-test to determine if the variances are equal:

import numpy as np
from scipy.stats import f

def generate_samples (mean1, var1, size1, mean2, var2, size2) :
  # Generate random samples from the two normal distributions
  sample1 = np.random.normal(mean1, np.sqrt(var1), size1)
  sample2 = np.random.normal(mean2, np.sqrt(var2), size2)
  return sample1, sample2

def variance_ratio_test (sample1, sample2) :
  # Calculate the variances of the samples
  var1 = np.var(sample1, ddof=1)
  var2 = np.var(sample2, ddof=1)

  # Calculate the F-value
  F = var1/var2 if var1 >= var2 else var2/var1

  # calculate the degrees of freedom for the two samples
  dfn = len(sample1) -1  # Degrees of freedom for the numerator
  dfd = len(sample2) -1  # Degrees of freedom the denominator

  # calculate the p-value
  p_value = 1- f.cdf (F, dfn, dfd)

  return F, dfn, dfd, p_value

# Parameters for the normal distributions

mean1 = 0
var1 = 1
size1 = 30

mean2 = 0
var2 = 1
size2 = 30

# Generate samples
sample1, sample2 = generate_samples (mean1, var1, size1, mean2, var2, size2)

# Perform the variance raion test
F, dfn, dfd , p_value = variance_ratio_test (sample1, sample2)

# Output the results
print("F-value:", F)
print("Degrees of freedom (numertor):", dfn)
print("Degrees of freedom (denominator):", dfd)
print("p-value:",p_value)


# Conclusion :
#You can modify the parameters for the normal distributions (mean, variance, and sample size) to test different scenarios.

```

```python
#Ans5.

#To test the manufacturer's claim about the variance of the diameter of a product using an F-test.

# State the Hypothesis :

# Null Hypothesis (Ho) :  The population variance σ^2 = 0.005
# Alternate Hypothesis (H1) : The population variance σ^2 ≠ 0.005

# Calculate the F-value :

# F = sample variance / claimed variance

 # Determine the critical F-values at the 1% significance level for a two-tailed test.
 # Compare the calculated F-value with the critical F-value to make a decision.

# Used Python Code to perform the F-test :

from scipy.stats import f

# Given variances and sample size
claimed_variance = 0.005
sample_variance = 0.006
n = 25

# Calculate the F-value
F = sample_variance / claimed_variance

# Degree of freedom for the sample
dfn = n-1  # Degree of freedom for the numerator
dfd = n-1  # Degree of freedom for the denominator (since it's a sigle sample)

# Significance level
alpha = 0.01

# Critical F-values for a two-tailed test
critical_value_lower = f.ppf(alpha / 2, dfn, dfd)
critical_value_upper = f.ppf(1-alpha/2, dfn, dfd)

# p-value for the calculated F-value
p_value = 2*(1 - f.cdf (F, dfn, dfd))  # two-tailed test

# output the results

print ("F-value:", F)
print ("Degrees of freedom:", dfn)
print ("critical F-value (lower):", critical_value_lower)
print ("critical F-value (upper):", critical_value_upper)
print ("p-value:", p_value)

# Decision

if F < critical_value_lower or F > critical_value_upper :
  print ("Reject the Null Hypothesis: The manufacturer's claim is not justified.")

else :
  print ("Fail to reject the Null Hypothesis : The manufacturer's claim is justified." )


# Conclusion :
 Running this code you can determine if the manufacturer's claim about the varianca is justified at the 1% significance level.RuntimeWarning

```

Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
F-distribution and calculates the mean and variance of the distribution. The function should return the
mean and variance as a tuple.

```python
#Ans6.

# To calculate the mean and variance of an F-distribution given the degree of freedom for the numerator(dfn), and denominator (dfd), you can use formulas :

# Mean  = ( d2 / d2 - 2 )   for  d2 > 2

# Variance = { 2d2^2(d1 + d2- 2) / d1(d2 - 2)^2(d2 - 4) }   for d2 > 4

# Python function to calculate the mean and variance of the F-distribution :


def f_distribution_mean_variance (dfn, dfd) :
  # Calculate the mean of the F-distribution
  if dfd > 2 :
    mean = dfd / (dfd - 2)
  else :
    mean = float('inf')  # Mean is undefined for dfd <= 2

  # Calculate the variance of the F-distribution
  if dfd > 4 :
    variance = (2 * (dfd ** 2) * (dfn + dfd - 2)) / (dfn * ((dfd - 2) **2) * (dfd - 2))
  else :
    variance = float ('inf')  # Variance id undefined for dfd <= 4

  return mean , variance

# Example usage

dfn = 5  # Degree of freedom for the numerator
dfd = 10 # Degree of freedom for the denominator
mean, variance = f_distribution_mean_variance (dfn, dfd)

print ("Mean:", mean)
print ("Variance:", variance)


# Conclusion :
#This function returns the mean and variance of the F-distribution as a tuple, handling cases where the mean and variance are undefined.UnicodeDecodeError

```

```python
#Ans8.

# Second method :  Used Python Code :-

import numpy as np
from scipy.stats import f

# Given data
restaurant_a = [24, 25, 28, 23, 22, 20, 27]
restaurant_b = [31, 33, 35, 30, 32, 36]

# Calculate the sample variances and sizes
var_a = np.var(restaurant_a, ddof=1)
var_b = np.var(restaurant_b, ddof=1)
n_a = len(restaurant_a)
n_b = len(restaurant_b)

# Calculate the F-value
F = var_a / var_b if var_a > var_b else var_b / var_a

# Degrees of freedom for the numerator and denominator
dfn = n_a - 1 if var_a > var_b else n_b - 1
dfd = n_b - 1 if var_a > var_b else n_a - 1

# Significance level
alpha = 0.05

# Critical F-values for a two-tailed test
critical_value_lower = f.ppf(alpha / 2, dfn, dfd)
critical_value_upper = f.ppf(1 - alpha / 2, dfn, dfd)

# p-value for the calculated F-value
p_value = 2 * (1 - f.cdf(F, dfn, dfd))  # Two-tailed test

# Output the results
print("Sample variance (Restaurant A):", var_a)
print("Sample variance (Restaurant B):", var_b)
print("F-value:", F)
print("Degrees of freedom (numerator):", dfn)
print("Degrees of freedom (denominator):", dfd)
print("Critical F-value (lower):", critical_value_lower)
print("Critical F-value (upper):", critical_value_upper)
print("p-value:", p_value)

# Decision
if F < critical_value_lower or F > critical_value_upper:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```
